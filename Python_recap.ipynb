# Metody/funcie/cykly

Napiste funkciu, ktora vrati mensie z 2 cisel ak su obe parne a vacsie ak nie.

```python
# Kod
def compare_two(a, b):
    if a % 2 == 0 and b % 2 == 0:
        return min(a, b)
    else:
        return max(a, b)

print(compare_two(4, 6))  # 4
print(compare_two(3, 6))  # 6
```

Napiste funkciu, ktora vrati True ak je sucet 2 zadanych cisel 30 alebo je jedno cislo 30, inak vrati False.

```python
#kod
def check_30(x, y):
    return x == 30 or y == 30 or (x + y) == 30

print(check_30(20, 10))  # True
print(check_30(30, 5))   # True
print(check_30(10, 5))   # False
```

Napiste funkciu ktora vrati True ak je zadane cislo vzdialene od cisla 100 alebo 200 maximalne o 10 .

```python
#Kod
def near_100_or_200(n):
    return abs(100 - n) <= 10 or abs(200 - n) <= 10

print(near_100_or_200(95))  # True
print(near_100_or_200(210)) # True
print(near_100_or_200(50))  # False
```

Napiste funkciu kde su zadane 3 cisla od 1 do 11, ak je ich sucet mensi rovny 21 vrati ich sucet ak je ich sucet vacsi ako 21 vrati ich sucet zmenseny o 10.

```python
#Kod
def sum_or_adjust(a, b, c):
    total = a + b + c
    if total > 21:
        return total - 10
    return total

print(sum_or_adjust(5, 6, 7))   # 18
print(sum_or_adjust(10, 7, 6))  # 13
print(sum_or_adjust(9, 8, 6))   # 13 (23-10)
```
