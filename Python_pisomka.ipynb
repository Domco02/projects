```python
import numpy as np

def matica(A, B):
    A_np = np.array(A)
    B_np = np.array(B)
    return A_np + B_np


A = [[2,2], [4,4]]
B = [[3,7], [2,5]]
C = matica(A, B)
print(C)
```

4. Definujte funkciu, ktorá vráti menšie z 2 čísel ak su obe párne a väčšie ak čísla párne niesu.

```python
def porovnaj(x, y):
    if x % 2 == 0 and y % 2 == 0:
        return min(x, y)
    else:
        return max(x, y)

print(porovnaj(9, 8))  
print(porovnaj(3, 5)) 
```
